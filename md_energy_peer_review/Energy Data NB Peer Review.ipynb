```python
#numbers are in million metric tons of CO2
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')

#Read CSVs
residential = pd.read_csv('residential.csv')

commercial = pd.read_csv('commercial.csv')

industrial = pd.read_csv('industrial.csv')

transportation = pd.read_csv('transportation.csv')

electricity = pd.read_csv('electricity.csv')


residential = residential.set_index('Source')
commercial = commercial.set_index('Source')
industrial = industrial.set_index('Source')
transportation = transportation.set_index('Source')
electricity = electricity.set_index('Source')

#for df in [residential, commercial, industrial, transportation, electricity]:
    #print(df)
```

```python
#Switches x and y axes of the dataframe
def switchaxes(df):
    df = df.transpose()
    return(df)

print("*Emissions numbers are in million metric tons of CO2\n")

residential_t = switchaxes(residential)
print(residential_t.head())

commercial_t = switchaxes(commercial)
print(commercial_t.head())

industrial_t = switchaxes(industrial)
print(industrial_t.head())

transportation_t = switchaxes(transportation)
print(transportation_t.head())

electricity_t = switchaxes(electricity)
print(electricity_t.head())
```

The next few cells will be graphing the emissions of each energy source (Coal, Petroleum, Natural Gas) in each sector (Residential, Commercial, Industrial, Transportation, and Electric Power) over time.

```python
residential_t.plot.line(y=['Coal','Petroleum Products','Natural Gas'], figsize = (8,6), title = "Maryland Residential Sector Emissions", xlabel="Year", ylabel="CO2 Emissions (million metric tons)")
plt.ylim(ymin=0,ymax=5.5)
```

```python
commercial_t.plot.line(y=['Coal','Petroleum Products','Natural Gas'], figsize = (8,6), title = "Maryland Commercial Sector Emissions", xlabel="Year", ylabel="CO2 Emissions (million metric tons)")
```

```python
industrial_t.plot.line(y=['Coal','Petroleum Products','Natural Gas'], figsize = (8,6), title = "Maryland Industrial Sector Emissions", xlabel="Year", ylabel="CO2 Emissions (million metric tons)")
```

```python
transportation_t.plot.line(y=['Coal','Petroleum Products','Natural Gas'], figsize = (8,6), title = "Maryland Transportation Sector Emissions", xlabel="Year", ylabel="CO2 Emissions (million metric tons)")
```

```python
electricity_t.plot.line(y=['Coal','Petroleum Products','Natural Gas'], figsize = (8,6), title = "Maryland Electricity Sector Emissions", xlabel="Year", ylabel="CO2 Emissions (million metric tons)")
```

### Analysis:
From these graphs, it is clear that in most of these sectors in Maryland, coal is not extensively used, and its use in the industrial and electricity generation sectors have been greatly reduced in recent years. Additionally, the presence of coal is also exaggerated from these graphs due to the plot being of carbon dioxide emissions, with coal generating much more CO2 compared to the other two sources. Therefore, especially in recent years, there is very little coal being used in Maryland.

In most sectors, since 1970-2019 petroleum products have been on the decline, and natural gas usage has risen. The only outlier is the transportation sector, since most vehicles run on petroleum products such as gasoline and diesel.

The next cell will graph the total emissions for all of the sources of each of the sectors for each year.

```python
plt.figure(figsize=(8,6))
plt.plot(residential_t.index, residential_t['Total'], label="Residential")
plt.plot(commercial_t.index, commercial_t['Total'], label="Commercial")
plt.plot(industrial_t.index, industrial_t['Total'], label="Industrial")
plt.plot(transportation_t.index, transportation_t['Total'], label="Transportation")
plt.plot(electricity_t.index, electricity_t['Total'], label="Electricity")
plt.xticks(range(0,len(residential_t.index),10))

#Add title and axes labels
plt.title("Total Emissions of Each Sector in Maryland")
plt.xlabel("Year")
plt.ylabel("CO2 Emissions (Million Metric Tons)")

#Add Legend
plt.legend(bbox_to_anchor=(1,1))
plt.show
```

### Analysis
The graph of the total emissions of each of the sectors shows us that initially, Industry, Transportation, and Electricity were responsible for most of the carbon emissions in Maryland, while Residential and Commercial emissions have both remained relatively low throughout. From 1970 to 2019, emissions from Industry have fallen steadily, from the highest emitting sector to the lowest. Similarly, while Electricity emissions rose from 1970 to around 2000, from that point on they also began to fall. This is most likely due to Maryland's new focus on introducing less emitting energy plants and deactivating the coal ones, which can also be seen from the previous graph of the individual sector. In contrast, Transportation emissions have slightly increased in the past half century, likely due to continuing increased demand for transport. Currently, the only major source of carbon emissions in Maryland is from transportation. Therefore, Maryland should begin focusing on implementing less polluting transportation options such as public transport and encouraging ownership of low emissions/hybrid/electric cars.

To aid in seeing how each sector and fuel source has contributed to carbon emissions in the past 50 years, the cell below will print the mean emissions of each sector as well as the mean emissions of each fuel source.

```python
#Lists & Variables
sector_mean = []
source_mean = []
labels = ["Residential","Commercial","Industrial","Transportation","Electricity"]


for df in [residential_t, commercial_t, industrial_t, transportation_t, electricity_t]:
    sector_mean.append(np.mean(df["Total"].tolist()))

#Print out the data
counter = 0
for item in sector_mean:
    print(labels[counter] + " Sector Average Yearly Emissions:\n" + str(round(item, 3)) + " Million Metric Tons\n")
    counter = counter + 1
    
#Finding the total of each energy source
coal = residential_t["Coal"]
petrol = residential_t["Petroleum Products"]
gas = residential_t["Natural Gas"]
for df in [commercial_t, industrial_t, transportation_t, electricity_t]:
    coal = coal + df["Coal"]
    petrol = petrol + df["Petroleum Products"]
    gas = gas + df["Natural Gas"]
    
#Print out the data
print("\nCoal Usage Average Yearly Emissions:\n" + str(round(np.mean(coal),3)) + " Million Metric Tons\n")
print("Petroleum Usage Average Yearly Emissions:\n" + str(round(np.mean(petrol),3)) + " Million Metric Tons\n")
print("Natural Gas Usage Average Yearly Emissions:\n" + str(round(np.mean(gas),3)) + " Million Metric Tons")
```

The next cell will graph the total emissions of all the sectors in Maryland.

```python
total = residential_t["Total"] + commercial_t["Total"] + industrial_t["Total"] + transportation_t["Total"] + electricity_t["Total"]
total.plot.line(figsize=(8,6), title="Total Carbon Emissions in Maryland", xlabel="Year", ylabel="CO2 Emissions (Million Metric Tons)")
```
